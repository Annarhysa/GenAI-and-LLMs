```python
from transformers import pipeline
summarizer = pipeline("summarization")
```

```python
from bs4 import BeautifulSoup as soup
from urllib.request import urlopen
from newspaper import Article
```

```python
import nltk
nltk.download('punkt')
```

```python
def fetch_news_search_topic(topic):
    site = 'https://news.google.com/rss/search?q={}'.format(topic)
    op = urlopen(site)  # Open that site
    rd = op.read()  # read data from site
    op.close()  # close the object
    sp_page = soup(rd, 'xml')  # scrapping data from site
    news_list = sp_page.find_all('item')  # finding news
    return news_list
```

```python
def fetch_top_news():
    site = 'https://news.google.com/news/rss'
    op = urlopen(site)  # Open that site
    rd = op.read()  # read data from site
    op.close()  # close the object
    sp_page = soup(rd, 'xml')  # scrapping data from site
    news_list = sp_page.find_all('item')  # finding news
    return news_list
```

```python
def fetch_category_news(topic):
    site = 'https://news.google.com/news/rss/headlines/section/topic/{}'.format(topic)
    op = urlopen(site)  # Open that site
    rd = op.read()  # read data from site
    op.close()  # close the object
    sp_page = soup(rd, 'xml')  # scrapping data from site
    news_list = sp_page.find_all('item')  # finding news
    return news_list
```

```python
def display_news(list_of_news, news_quantity):
    c = 0
    for news in list_of_news:
        c += 1
        print(c, news.title.text)
        try:
          data = Article(news.link.text)
          data.download()
          data.parse()
          data.nlp()
          final_data = summarizer(data.summary)
          print(final_data[0]['summary_text'])

        except:
          print("Couldn't fetch data")

        if c >= news_quantity:
            break
```

```python
category = ['Trending News', 'Favourite Topics', 'Search Topic']
```

```python
topics = ['WORLD', 'NATION', 'BUSINESS', 'TECHNOLOGY', 'ENTERTAINMENT', 'SPORTS', 'SCIENCE','HEALTH']
```

```python
def choices(category_choice):
  if category_choice == category[0]:
    news_list = fetch_top_news()
  elif category_choice == category[1]:
    print(topics)
    topic_choice = input("Choose from the above ")
    news_list = fetch_category_news(topic_choice)
  elif category_choice == category[2]:
    input_choice = input("Enter a topic you want to fetch news about ")
    news_list = fetch_news_search_topic(input_choice)
  else:
    print("No News Found")

  return news_list
```

```python
print(category)
category_choice = input("Choose from the above ")
```

```python
news_list = choices(category_choice)
display_news(news_list, 2)
```
